```python
# 2.9) создание нового поля вычисленного на основе значений других полей:
# 2.9.1) через выражение на базе имеющихся колонок
url = 'https://archive.ics.uci.edu/static/public/53/data.csv'
column_names = ['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка', 'Класс']
df3 = pd.read_csv(url, header=None, names=column_names)
df3.columns = column_names
df3.drop (index=df3.index [0], axis= 0 , inplace= True )
df3[['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']]=df3[['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']].astype(float)
df3['Площадь чашелистика'] = df3['Длина чашелистика'] * df3['Ширина чашелистика']
df3['Площадь лепестка'] = df3['Длина лепестка'] * df3['Ширина лепестка']
df3
```

```python
# 2.9.2) через DataFrame.apply
url = 'https://archive.ics.uci.edu/static/public/53/data.csv'
column_names = ['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка', 'Класс']
df4 = pd.read_csv(url, header=None, names=column_names)
df4.columns = column_names
df4.drop (index=df4.index [0], axis= 0 , inplace= True )
df4[['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']]=df4[['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']].astype(float)
df4['Площадь чашелистика'] = df4['Длина чашелистика'] * df4['Ширина чашелистика']
def get_sepal_size(row):
    if row['Площадь чашелистика'] < 18:
        return 'Малый'
    else:
        return 'Большой'
df4['Размер чашелистика'] = df4.apply(get_sepal_size, axis=1)
df4
```

```python
# 2.9.3) через Series.apply
url = 'https://archive.ics.uci.edu/static/public/53/data.csv'
column_names = ['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка', 'Класс']
df5 = pd.read_csv(url, header=None, names=column_names)
df5.columns = column_names
df5 = df5.rename(columns={'class':'iris_sc_name'})
df5.drop (index=df5.index [0], axis= 0 , inplace= True )
name = list(df5['Класс'])
ser = pd.Series(name)
def get_name(iris_sc_name):
    if iris_sc_name == 'Iris-setosa':
        return 'Ирис щетинистый'
    elif iris_sc_name == 'Iris-versicolor':
        return 'Ирис версиколор'
    else:
        return 'Ирис виргинский'
ser.apply(get_name)
```

```python
# 2.10) сортировка по какому-либо из полей
# 2.11) вычислить несколько статистик по колонкам.
url = 'https://archive.ics.uci.edu/static/public/53/data.csv'
column_names = ['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка', 'Класс']
df6 = pd.read_csv(url, header=None, names=column_names)
df6.columns = column_names
df6.drop (index=df6.index [0], axis= 0 , inplace= True )
df6[['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']]=df6[['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']].astype(float)
df6['Площадь чашелистика'] = df6['Длина чашелистика'] * df6['Ширина чашелистика']
mean_petal_area = df6['Площадь чашелистика'].mean()
min_petal_area = df6['Площадь чашелистика'].min()
max_petal_area = df6['Площадь чашелистика'].max()
print("Среднее значение: ", mean_petal_area)
print("Минимальное значение: ", min_petal_area)
print("Максимальное значение: ", max_petal_area)
```
